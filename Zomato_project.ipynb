```python
data['rate']=data['rate'].str.replace('/5','')
data['rate']=data['rate'].astype(float)
```

```python
ax =sns.countplot(data['online_order'],palette=sns.color_palette('pastel'))
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
for container in ax.containers:
    ax.bar_label(container)
plt.title('Online Order Distribution')
plt.xlabel('Online Order')
plt.ylabel('Count')
plt.legend()
```

```python
ax = sns.histplot(data['rate'],kde=True)
for container in ax.containers:
    ax.bar_label(container)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.legend()
```
